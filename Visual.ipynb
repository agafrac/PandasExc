```python
bb["Close"]
```

```python
def rank_perf(stack_price):
    if stack_price <= 10:
        return "Low"
    elif stack_price <= 50:
        return "Good"
    else:
        return "Very_good"
```

```python
bb["Close"].apply(rank_perf).value_counts().plot(kind = "barh")
```

```python
m = bb["Open"].mean()
```

```python
def rank2(stock_price):
    if stock_price >= m:
        return "above"
    else:
        return "below"
```

```python
bb["Open"].apply(rank2).value_counts().plot(kind = "pie", legend = True)
```
