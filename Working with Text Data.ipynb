```python
import pandas as pd
```

```python
chicago = pd.read_csv("chicago.csv")
chicago.head(3)
```

```python
chicago[["New Name", "New"]] = chicago["Name"].str.split(",", expand = True)
chicago.head(3)
```

```python
chicago.columns = chicago.columns.str.replace(" ", "_")
chicago.head(3)
```

```python
chicago = pd.read_csv("chicago.csv", index_col = "Name")
chicago.head(3)
```
